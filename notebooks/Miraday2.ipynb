```python
#max temperature time series
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
ds_solar = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6solar/r1i1p1f2/Amon/tasmax/gn/files/d20191031/*')
ds_sulfur = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6sulfur/r1i1p1f2/Amon/tasmax/gn/files/d20191113/*')
ds_245 = xr.open_mfdataset('/badc/cmip6/data/CMIP6/ScenarioMIP/MOHC/UKESM1-0-LL/ssp245/r1i1p1f2/Amon/tasmax/gn/files/d20190715/*')
ds_585 = xr.open_mfdataset('/badc/cmip6/data/CMIP6/ScenarioMIP/MOHC/UKESM1-0-LL/ssp585/r1i1p1f2/Amon/tasmax/gn/files/d20190726/*')
pi_control = xr.open_mfdataset('/badc/cmip6/data/CMIP6/CMIP/MOHC/UKESM1-0-LL/piControl/r1i1p1f2/Amon/tasmax/gn/files/d20190827/*').isel(time=slice(0,1200))

ds_solar_tropics = ds_solar.tasmax.where(ds_solar.lat >-20).where(ds_solar.lat <20).mean(('lat','lon'))
ds_sulfur_tropics = ds_sulfur.tasmax.where(ds_sulfur.lat >-20).where(ds_sulfur.lat <20).mean(('lat','lon'))
ds_245_tropics = ds_245.tasmax.where(ds_245.lat >-20).where(ds_245.lat <20).mean(('lat','lon'))
ds_585_tropics = ds_585.tasmax.where(ds_585.lat >-20).where(ds_585.lat <20).mean(('lat','lon'))
pi_ = pi_control.tasmax.where(pi_control.lat >-20).where(pi_control.lat <20).mean(('lat','lon'))

plt.figure(figsize=(8, 4),dpi=150)
plt.plot((ds_solar_tropics-273).groupby('time.year').max('time').values,label='G6 Solar')
plt.plot((ds_sulfur_tropics-273).groupby('time.year').max('time').values,label='G6 Sulfur')
plt.plot((ds_245_tropics-273).groupby('time.year').max('time').values,label='SSP 245')
plt.plot((ds_585_tropics-273).groupby('time.year').max('time').values,label='SSP 585')
plt.plot((pi_-273).groupby('time.year').max('time').values,color = 'black', label='Preindustrial Control')

plt.grid()
#plt.ylim([14,22])
plt.ylabel('$^\circ$C')
plt.xlabel('Years')
plt.legend()
plt.title('Tropics Yearly Maximum Near-Surface Air Temperature (tasmax) 20S-20N')
plt.show()
```

```python
#sulfur - solar difference timeseries for both min and max temperatures

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
ds_solar = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6solar/r1i1p1f2/Amon/tasmin/gn/files/d20191031/*')
ds_sulfur = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6sulfur/r1i1p1f2/Amon/tasmin/gn/files/d20191113/*')

maxds_solar = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6solar/r1i1p1f2/Amon/tasmax/gn/files/d20191031/*')
maxds_sulfur = xr.open_mfdataset('/badc/cmip6/data/CMIP6/GeoMIP/MOHC/UKESM1-0-LL/G6sulfur/r1i1p1f2/Amon/tasmax/gn/files/d20191113/*')

ds_solar_tropics = ds_solar.tasmin.where(ds_solar.lat >-20).where(ds_solar.lat <20).mean(('lat','lon'))
ds_sulfur_tropics = ds_sulfur.tasmin.where(ds_sulfur.lat >-20).where(ds_sulfur.lat <20).mean(('lat','lon'))

maxds_solar_tropics = maxds_solar.tasmax.where(maxds_solar.lat >-20).where(maxds_solar.lat <20).mean(('lat','lon'))
maxds_sulfur_tropics = maxds_sulfur.tasmax.where(maxds_sulfur.lat >-20).where(maxds_sulfur.lat <20).mean(('lat','lon'))

diffmin= ds_sulfur_tropics - ds_solar_tropics
diffmax= maxds_sulfur_tropics - maxds_solar_tropics

plt.figure(figsize=(8, 4),dpi=150)
plt.plot((diffmin).groupby('time.year').min('time').values,label='sulfur-solar difference tasmin')
plt.plot((diffmax).groupby('time.year').min('time').values,label='sulfur-solar difference tasmax')

plt.grid()
plt.ylabel('$^\circ$C')
plt.xlabel('Years')
plt.legend()
plt.title('Difference between G6 sulfur and G6 solar (sulfur-solar)')
plt.show()
```
